```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
%matplotlib inline 
```

Load train and test data


```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train.describe()
```

```python
numColumns = ['MSSubClass',]
xColumns = ['LotArea', 'OverallQual','OverallCond']
yColumns = ['SalePrice']

#split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(train[xColumns], train[yColumns], test_size=0.33, random_state=42)

#print(X_train.shape,y_train.shape)

#train.hist(figsize=(20,20))
```

We try to fit a linear regression model through the dataset.

```python
# Create linear regression object
regr = LinearRegression(normalize=True,copy_X=True)

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))

# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test,y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

We saw the basic summary of data. We also saw the distribution of data through histogram.
