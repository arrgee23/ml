```python
def train_test_split(df,test_prop):
    test_size = round(test_prop*len(df))
    
    indices = df.index.tolist()
    
    test_indices = random.sample(population=indices,k=test_size)
    
    test_df = df.loc[test_indices]
    train_df = df.drop(test_indices)
    
    return train_df,test_df
```

```python
# checks if the given array of data is pure
def check_purity(data):
    label_column = data[:,-1]
    if(len(np.unique(label_column)) == 1):
        return True
    else:
        return False
```

```python
def classify_data(data):
    label_column = data[:,-1]
    unique_classes, unique_counts = np.unique(label_column, return_counts=True)
    index = unique_counts.argmax()
    maximal_class = unique_classes[index]
    
    return maximal_class
```

```python
""" 
 make a dictionary of potential splits
 key is the index of the column
 value is a list of all potential split values for that column
 for example the 3rd col {2:[1.4,1.3,1.5,1.7]}
"""
def get_potential_splits(data):
    potential_splits = {}
    n_cols = data.shape[1]

    for i in range(n_cols-1):
        potential_splits[i] = []
        unique_values = np.unique(data[:,i])
        #print(unique_values)
        potential_splits[i] = unique_values

        #print(potential_splits)
    return potential_splits
```

### Split Data

```python
"""Given a splitpoint splits the np table into two parts and returns 
them"""
def split_data(data,split_column,split_value):
    
    split_col_values = data[:,split_column]
    data_below = data[split_col_values<=split_value]
    data_above = data[split_col_values>split_value]
    
    return data_below,data_above
```

```python
split_column = 3
split_value  = 1.05
data_below,data_above = split_data(data,split_column,split_value)
```

### Entropy

```python
def calculate_entropy(data):
    label_column = data[:,-1]
    _,counts = np.unique(label_column,return_counts=True)

    probabilities = counts/counts.sum()
    entropy = sum(probabilities * (-np.log2(probabilities)))
    return entropy
```

```python
entropy = calculate_entropy(data_below)
entropy 
```

```python
"""
    given a split calculate the overall entropy
"""
def calculate_overall_entropy(data_below,data_above):
    p_data_below = len(data_below)/(len(data_below)+len(data_above))
    p_data_above = len(data_above)/(len(data_below)+len(data_above))   
    
    overall_entropy = p_data_below*calculate_entropy(data_below)+p_data_above*calculate_entropy(data_above)
    
    return overall_entropy
```

```python
calculate_overall_entropy(data_below,data_above)
```

```python

def determine_best_split(data,potential_splits):
    """
    Determine the split that produces minimum entropy
    """
    overall_entropy = 99999
    
    for col in potential_splits:
        for split_value in potential_splits[col]:
            data_below,data_above = split_data(data,col,split_value)
            new_entropy = calculate_overall_entropy(data_below,data_above)
        
            if(new_entropy<=overall_entropy):
                overall_entropy = new_entropy
                best_split_col = col
                best_split_val = split_value
            
    return best_split_col,best_split_val
```

```python
determine_best_split(data,potential_splits)
```

## Structure of the tree



subtree = {question : [yes_answer,no_answer]}

### Algo

```python
def decision_tree_algorithm(df,counter=0):
    
    #1st time we convert df to np table
    if counter == 0:
        data = df.values
    else:
        data = df
        
    #base case
    if check_purity(data):
        classification = classify_data(data)
        return classification
    
    else:
        counter += 1
        
        potential_splits = get_potential_splits(data)
        split_column,split_value = determine_best_split(data,potential_splits)
        
        data_below,data_above = split_data(data,split_column,split_value)
        
        question = "{} <= {}".format(split_column,split_value)
        
        sub_tree = {question: []}
        
        #find answers
        yes_answer = decision_tree_algorithm(data_below,counter)
        no_answer = decision_tree_algorithm(data_above,counter)
        
        sub_tree[question].append(yes_answer)
        sub_tree[question].append(no_answer)
        
        return sub_tree
        
```

```python
def parse_label(string):
    if(string[0]=='I'):
        return None,string
    else:
        words = string.split(' ')
        return int(words[0]),float(words[2])
```

```python
parse_label('Int')
```

```python
def evaluate_one_example(example,tree):
    if(type(tree)== str):
        return tree
    else:
        for key in tree:
            colIdx,value = parse_label(key)
            
            if(colIdx == None):# we have a pure label
                return value
            else: 
                if(example[colIdx]<= value): # left subtree
                    return evaluate_one_example(example,tree[key][0])
                else: # right subtree
                    return evaluate_one_example(example,tree[key][1])
            
```

```python
def evaluate_all_examples(test_df,tree):
    correct = 0
    incorrect = 0
    for i in range(len(test_df.values)):
        predicted_label = evaluate_one_example(test_df.values[i],tree)
        actual_label = test_df.values[i][-1]
        if(predicted_label == actual_label):
            correct += 1
        else:
            incorrect += 1
    return correct,incorrect
```
